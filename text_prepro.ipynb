```python
import re
def extract_ingredients(line):
    # Extract only the food item from the recipe ingredient
    ing_list_raw = line.split("[")[1].replace(']', '').replace("'", "").replace('"', '').split(',')
    ing_list_white_remed = [x.strip().lower() for x in ing_list_raw]
    ing_list_new = []
    for i in ing_list_white_remed:
        stringy = i
        regex = re.compile(".*?\((.*?)\)")
        result = re.findall(regex, stringy)
        for comb in result:
            stringy = stringy.replace(comb,'').replace('()', '').strip()
        stringy = re.sub('[a-z]*[®,™]', '', stringy).strip()
        resultwords = [word for word in re.split("\W+",stringy) if word not in stop_words]
        resultwords = [word for word in resultwords if word!='' and word not in stop_words]
        stringy = ' '.join(resultwords)
        if stringy!='' and stringy not in stop_words:
            ing_list_new.append(stringy)

    return ing_list_new
```

```python
title = []
total = []
ing = []
new_ing = []
old_title = ''
img_count = 0
# Creates list of files names and a list of their ingredients
with open("Recipes.txt","r", encoding='utf-8') as filename:
    for line in filename:
        title_ = line.split("[")[0].split(",")[0]
        if old_title!=title_:
            img_count=0
        old_title=title_
        img_path = title_+f'\\{img_count}.jpg'
        img_count+=1
        ll = line
        ingreds = extract_ingredients(ll)
        title.append(img_path)
        ing.append(ingreds)
```
